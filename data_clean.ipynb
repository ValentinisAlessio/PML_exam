CARD: cartellino. IDEM a sopra.
BALL OUT: palla fuori. Controlla SET PIECE subito dopo.



```python
index_list = events.loc[events['Type'] == 'SET PIECE'].index.to_list()
# slices_to_remove = [index_list[i] -1, index_list[i] for i in range(1, len(index_list))]
index_list.extend([index_list[i] -1 for i in range(1, len(index_list))])
# print(index_list)
# slices_to_remove
# Print the slices to remove
# events.iloc[slices_to_remove[:]:slices_to_remove[:]].head()
sorted_index_list = sorted(index_list)[1:]
print(sorted_index_list)
```

```python
events.iloc[sorted_index_list].head()
```

```python
frames_to_remove = []
for i in sorted_index_list[::2]:
    # print(events['End Frame'][i])
    # print(events['Start Frame'][i+1])
    frames_to_remove.extend(list(range(events['End Frame'][i], events['Start Frame'][i+1])))

print(len(frames_to_remove))
```

```python
card_list = events.loc[events['Type'] == 'CARD'].index.to_list()
card_list.extend([card_list[i] -1 for i in range(0, len(card_list))])
sorted_card_list = sorted(card_list)[:]
print(sorted_card_list)
```

```python
for i in sorted_card_list[::2]:
    # print(events['End Frame'][i])
    # print(events['Start Frame'][i+1])
    frames_to_remove.extend(list(range(events['End Frame'][i], events['Start Frame'][i+1])))

print(len(frames_to_remove))
```
